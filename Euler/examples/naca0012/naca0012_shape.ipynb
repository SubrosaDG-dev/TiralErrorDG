```julia
import DelimitedFiles;
import PyPlot as plt;
rcParams = plt.PyDict(plt.matplotlib."rcParams");
rcParams["text.usetex"] = true;
```

```julia
naca0012_file::String = "naca0012.dat";
naca0012_data::Matrix{Float64} = DelimitedFiles.readdlm(naca0012_file; skipstart=1)[:, 1: 2];
```

```julia
plt.figure(figsize=(10, 4));
plt.plot(naca0012_data[:, 1], naca0012_data[:, 2], color="k", linestyle="-", linewidth=1);
plt.scatter(naca0012_data[:, 1], naca0012_data[:, 2], color="r", s=10, alpha=0.6, zorder=10);
plt.xlabel("\$\\frac{x}{c}\$", fontsize=20);
plt.ylabel("\$\\frac{y}{c}\$", fontsize=20);
plt.axis("equal");
plt.grid();
display(plt.gcf());
plt.show();
```
